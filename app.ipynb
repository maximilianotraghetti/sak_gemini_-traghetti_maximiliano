# TRABAJO PRACTICO UTILIZACION SDK GEMINI

```python
from google import genai #importamos la libreria genai

import os #importamos la libreria os

load_dotenv() #cargamos las variables de entorno desde el archivo .env

#creamos un cliente de genai con la API_KEY
client = genai.Client(api_key="Aca tu api key")
#generamos contenido con el modelo gemini-2.5-flash
response = client.models.generate_content(
    model= "gemini-2.5-flash",
    contents="¿Cuales son los primeros 8 digitos de pi?"
)

print(f"Respuesta: {response.text}") 
```

Se observa un importante latencia a la hora de que el modelo arroje una respuesta.

```python
from google import genai #importamos la libreria genai
import os #importamos la libreria os

#creamos un cliente de genai con la API_KEY
client = genai.Client(api_key="Aca tu api key")

#generamos contenido con el modelo gemini-25-flash (streaming)
response = client.models.generate_content_stream(
    model= "gemini-2.5-flash",  # ⬅️ Modelo disponible (cambiado de 2.5 a 1.5)
    contents="¿Cuales son los primeros 8 digitos de pi?",
)

for chunk in response: #iteramos sobre la respuesta
    print(chunk.text, end="", flush=True) #imprimimos cada fragmento de la respuesta
```

En comparacion el primer bloque de codigo , utilizando el generate_content_stream , se acorta mucho el tiempo de respuesta.

```python
from google.genai import types

YOUR_IMAGE_PATH = './barista_preparando_cafe.jpg'
YOUR_IMAGE_MIME_TYPE = 'image/jpeg'
with open(YOUR_IMAGE_PATH, 'rb') as f:
    image_bytes = f.read()

for chunk in client.models.generate_content_stream(
    model='gemini-2.0-flash-001',
    contents=[
        'Describe a la persona con anteojos en la imagen y que esta haciendo.',
        types.Part.from_bytes(data=image_bytes,
        mime_type=YOUR_IMAGE_MIME_TYPE),
],
):
    print(chunk.text, end='')
```

En base a la imagen analizada , se puede ver que la respuesta arrojada por el modelo es bastante correcta.
